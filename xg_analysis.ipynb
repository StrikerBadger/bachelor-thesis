```python
from matplotlib import pyplot as plt
import pandas as pd
import math
```

```python
# Load the data
xg_df = pd.read_pickle('dataset/intersection/statsbomb_matches_shots/shots_per_match.pkl')

xg_df.head()
```

```python
# Plot histogram of the xGs
xgs = xg_df['shot_xgs'].sum()

plt.hist(xgs, bins=math.ceil(math.sqrt(len(xgs))))
plt.title('Histogram of xG values')
plt.xlabel('xG')
plt.ylabel('Number of shots')
plt.show()
```

```python
# Plot histogram of the shots per team
plt.hist(xg_df['shot_amnt'], bins=xg_df['shot_amnt'].max())
plt.title('Histogram of shots per team')
plt.xlabel('Amount of shots')
plt.ylabel('Number of matches')
plt.show()
```
